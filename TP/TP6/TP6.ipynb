# **1. Utilisateurs :**

### **1. Mettre en place le formulaire de création d'un utilisateur, il faudra vérifier que le compte n'existe pas :**

On commence par créer la vue pour la création d'utilisateur dans `views.py`, on la nomme `creer`. On peut s'inspirer grandement du code de `edit_profile`, si on accède à la vue par la méthode get, on créé un formulaire vide et on le passe en paramètre. Si au contraire on y accède par la méthode post, alors l'utilisateur a déjà rempli le formulaire, on récupère alors les données envoyées. On vérifie ensuite si les données son correctes et si l'utilisateur existe déjà dans la base de données. Si tout est bon, on créé l'utilisateur et on affiche un message de confirmation, sinon, on n'ajoute rien dans la base et affiche une erreur.

On ajoute ensuite un lien vers la vue dans `urls.py`, puis on rajoute un lien vers la page de création de profil dans la template `base.html`. Pour terminer, on créé la template de création d'utilisateur, `create_profile.html`.

### **2. Permettre à l'utilisateur de modifier son mot de passe :**

On commence par créer la vue pour le changement de mot de passe dans `views.py`, on la nomme `change`, on ajoute à cette vue un décorateur obligeant l'utilisateur à être connecté pour effectuer cette opération. Si on accède à la vue par la méthode get, on créé un formulaire vide et on le passe en paramètre. Si au contraire on y accède par la méthode post, alors l'utilisateur a déjà rempli le formulaire, on récupère alors les données envoyées. On vérifie ensuite si les données son correctes. Si tout est bon, on change le mot de passe de l'utilisateur actuellement en session pour mettre le nouveau mot de passe demandé. Sinon on ne change rien et on affiche une erreur.

On ajoute ensuite un lien vers la vue dans `urls.py`, puis on rajoute un lien vers la page de changement de mot de passe dans la template `base.html`. Pour terminer, on créé la template de changement de mot de passe, `change_password.html`.

Après un changement de mot de passe, l'utilisateur se retrouve déconnecté car ses informations de connexion ne sont plus correctes (le mot de passe a changé), il faut donc se reconnecter.

### **3. Permettre à l'utilisateur de supprimer son compte :**

On commence par créer la vue pour la suppression de compte dans `views.py`, on la nomme `delete`, on ajoute à cette vue un décorateur obligeant l'utilisateur à être connecté pour effectuer cette opération. Si on accède à la vue par la méthode get, on créé un formulaire vide et on le passe en paramètre. Ce formulaire contient uniquement un champ mot de passe, on demande à l'utilisateur d'entrer son mot de passe en guise de confirmation pour supprimer son compte. Si au contraire on y accède par la méthode post, alors l'utilisateur a déjà rempli le formulaire, on récupère alors les données envoyées. On vérifie ensuite si les données son correctes. Si tout est bon, on vérifie ensuite que la chaîne entrée correspond bien au mot de passe de l'utilisateur. S'il y a un problème à une étape, on affiche une erreur, sinon, on supprime l'utilisateur de la base et on redirige vers une page de confirmation.

On ajoute ensuite un lien vers la vue dans `urls.py`, puis on rajoute un lien vers la page de suppression de compte dans la template `base.html`. Pour terminer, on créé la template de suppression de compte, `delete_account.html`.

# **2. Simulations :**

### **1. Permettre à un utilisateur de partager une simulation avec un autre utilisateur :**

Avant toute chose, on a besoin de créer un nouveau modèle. On créé donc un modèle dans `models.py`, qu'on nomme "Partage" et qui représentera des partage de simulations. On créé trois attributs, tous clé étrangères, un contenant l'utilisateur qui partage la simulation, un autre contenant l'utilisateur qui reçoit la simulation, et un dernier qui contient directement la simulation.

Pour partager une simulation avec un autre utilisateur, au lieu de créer une nouvelle vue, nous avons décidé de modifier la vue affichant la liste des simulations, `simulation_list`.
    
Pour cela on ajoute à la page un formulaire contenant tous les utilisateurs contenus dans la base. Si on accède à la page par get, on créé juste un formulaire vide et on le passe en paramètre, si on y accède avec post, alors on récupère l'identifiant de l'utilisateur qui partage la simulation (donc celui connecté), et l'identifiant de l'utilisateur à qui on envoie la simulation (donc l'utilisateur entré dans le formulaire). à cela on rajoute une troisième information, dans la template dans le formulaire, nous mettons un attribut caché (`input` avec comme `type hidden`), contenant en valeur l'identifiant de la simulation, pour chaque simulation dans la liste. On récupère donc cet attribut et on a l'identifiant de l'expéditeur, du destinataire et de la simulation. On peut donc créer une nouvelle instance de la classe "Partage", que nous enregistrons dans la base de données. On modifie donc la template de liste de simulations.

Ensuite, il faut avoir accès aux différentes simulations qui ont été partagées. On créé donc une nouvelle vue, `partage`, à laquelle on ajoute le décorateur obligeant l'utilisateur à être connecté. On récupère ensuite toutes les simulations envoyées et reçues par l'utilisateur et on les passe en paramètre.

On ajoute ensuite un lien vers cette vue dans `urls.py`, on rajoute un lien vers la page de liste des partages dans la template `base.html`, et pour terminer, on créé la template de liste des partages, `partage.html`.

### **2. Ajouter un système de favoris sur les simulations. Les simulations favoris se placeront en haut de la liste :**

Pour ajouter un système de favoris sur les simulations, il n'est pas nécessaire de créer une nouvelle classe dans `models.py`, on peut se contenter de rajouter un attribut de type booléen à la classe `Simulation`.

Pour gérer les favoris, on peut ici aussi le faire directement depuis la vue de liste des simulations, `simulation_list`. Comme il est demandé de placer les simulations favoris en haut de la liste, on récupère alors la liste de simulation en deux fois, avec des filtres. On récupère dans un premier temps les simulations de l'utilisateur connecté qu'il a mises en favori, puis les simulations de l'utilisateur qu'il n'a pas mises en favori. Si on accède à la vue via la méthode get, on renvoie alors les deux listes de simulations favori et non favori. Si on accède à la vue depuis la méthode post, alors on récupère la simulation en question grâce à un attribut caché (`input` de `type hidden`), si la simulation est favori alors on enlève le favori, si la simulations n'est pas en favori, alors on ajoute le favori. On recharge ensuite la page une fois la modification faite.

On modifie ensuite la template de liste de simulations pour afficher d'abord les simulations favoris puis ensuite les simulations non favoris. On ajoute aussi un bouton permettant de mettre et d'enlever le favori.

# **/!\ /!\ /!\ Attention /!\ /!\ /!\ :**

Les favoris apparaissent avec un bouton vert et les non favoris avec un bouton rouge. Lorsqu'on appuie sur un des boutons, on peut parfois avoir l'impression que c'est une autre simulation qui change de couleur, c'est juste une illusion. En effet les favoris apparaissent en premier dans la liste, et les non favoris en derniers, si vous ajouter un favori sur une simulation tout en bas de la liste, cette dernière va donc remonter et changer de place instantanément, donnant donc l'impression que c'est une autre simulation qui a changé de couleur. Pour vérifier que c'est bien la bonne simulation qui a été affectée, l'identifiant de chaque simulation est affiché dans la liste des simulations.
